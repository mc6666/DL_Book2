# 積分(Integration)

## 範例1. 積分
### $\int x\;dx = \frac{1}{2} x^2$

```python
# 載入套件
import numpy as np
import scipy.integrate as integrate
import numpy as np
import math

f = lambda x: x 
i, e = integrate.quad(f, 0, 3)

print('積分值: ' + str(i))
print('誤差: ' + str(e))
```

```python
import matplotlib.pyplot as plt

# 樣本點
x = range(0, 11)

# 作圖
plt.figure(figsize=(6,4))
plt.plot(x, f(x), color='purple')

# 積分面積
area = np.arange(0, 3, 1/20)
plt.fill_between(area, f(area), color='green')

# 設定圖形屬性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

plt.show()
```
